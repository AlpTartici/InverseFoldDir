```python
def correct_structure_with_b_factors(structure_id, entry, pdb_processor):
    """
    Process a single structure entry and add B-factors.
    
    Args:
        structure_id: ID of the structure (e.g., "3nks.A")
        entry: Original CATH entry dictionary
        pdb_processor: PDBProcessor instance
    
    Returns:
        Updated entry with B-factors added
    """
    # Create a copy of the original entry
    new_entry = entry.copy()
    coords_N = entry['coords']['N']
    
    # Get sequence length
    sequence = entry.get('seq', '')
    seq_length = len(sequence)
    
    # Extract B-factors for this structure
    # Convert structure ID format: "3nks.A" -> "3nks_chainA"
    if '.' in structure_id:
        pdb_id, chain = structure_id.split('.')
        structure_filename = f"{pdb_id}_chain{chain}"
    else:
        # Handle alternative formats
        structure_filename = structure_id
    
    try:
        b_factors = pdb_processor.get_b_factors_for_structure(structure_filename, seq_length)
        b_factors_aligned = np.ones(len(coords_N))*100

        nonNaN_pos = 0
        # Align B-factors with coordinates
        for i, coord in enumerate(coords_N):
            # Find the closest residue in B-factors
            if not coord[0] > -1000:
                continue
            b_factors_aligned[i] = b_factors[nonNaN_pos]
            nonNaN_pos += 1

        # Store raw B-factors (no normalization)
        new_entry['b_factors'] = b_factors_aligned.astype(np.float32)

        return new_entry, True

    except Exception as e:
        print(f"Warning: Failed to extract B-factors for {structure_id}: {e}")
        # Add default B-factors (50.0 = typical default value)
        new_entry['b_factors'] = np.full(seq_length, 50.0, dtype=np.float32)
        return new_entry, False
```

```python
# Independent correctness test: Verify B-factor extraction and averaging
print("\n" + "="*80)
print("INDEPENDENT CORRECTNESS TEST")
print("="*80)
print("Verifying that B-factor extraction and per-residue averaging is correct...")

def extract_bfactors_directly_from_pdb(pdb_file_path):
    """
    Extract B-factors directly from PDB file, exactly matching the PDBProcessor logic.
    This is an independent implementation for verification.
    """
    try:
        with open(pdb_file_path, 'r') as f:
            lines = f.readlines()
    except Exception as e:
        return None, f"Error reading file: {e}"
    
    # Extract B-factors from ATOM records (matching PDBProcessor logic)
    residue_b_factors = {}  # residue_id -> list of b_factors
    
    for line in lines:
        if line.startswith('ATOM'):
            try:
                # Parse PDB ATOM record (same as PDBProcessor)
                atom_name = line[12:16].strip()
                res_seq = int(line[22:26].strip())
                chain_id = line[21]
                b_factor = float(line[60:66].strip())
                
                # Create unique residue identifier
                res_id = f"{chain_id}_{res_seq}"
                
                # Only consider backbone atoms for representative B-factor (same as PDBProcessor)
                if atom_name in ['CA', 'C', 'N', 'O']:
                    if res_id not in residue_b_factors:
                        residue_b_factors[res_id] = []
                    residue_b_factors[res_id].append(b_factor)
            
            except (ValueError, IndexError) as e:
                continue
    
    # Average B-factors per residue (same as PDBProcessor)
    if residue_b_factors:
        avg_b_factors = []
        residue_details = []
        
        for res_id in sorted(residue_b_factors.keys()):
            avg_b_factor = np.mean(residue_b_factors[res_id])
            avg_b_factors.append(avg_b_factor)
            residue_details.append({
                'res_id': res_id,
                'num_atoms': len(residue_b_factors[res_id]),
                'atom_bfactors': residue_b_factors[res_id],
                'avg_bfactor': avg_b_factor
            })
        
        return np.array(avg_b_factors), residue_details
    
    return None, "No B-factors found"

# Test on a few structures
test_structures = list(new_cath_data.keys())[:3]
print(f"Testing {len(test_structures)} structures...")

for i, struct_id in enumerate(test_structures):
    print(f"\n{'='*60}")
    print(f"TEST {i+1}/3: {struct_id}")
    print(f"{'='*60}")
    
    # Get data from pickle file
    pickle_entry = new_cath_data[struct_id]
    pickle_bfactors = pickle_entry['b_factors']
    sequence = pickle_entry['seq']
    
    # Convert structure ID to PDB filename
    if '.' in struct_id:
        pdb_id, chain = struct_id.split('.')
        pdb_filename = f"{pdb_id}_chain{chain}.pdb"
    else:
        pdb_filename = f"{struct_id}.pdb"
    
    pdb_file_path = Path(pdb_directory) / pdb_filename
    
    print(f"Structure ID: {struct_id}")
    print(f"PDB file: {pdb_filename}")
    print(f"Sequence length: {len(sequence)}")
    print(f"Pickle B-factors shape: {pickle_bfactors.shape}")
    print(f"PDB file exists: {pdb_file_path.exists()}")
    
    if not pdb_file_path.exists():
        print(f"❌ PDB file not found, skipping...")
        continue
    
    # Extract B-factors directly from PDB
    direct_bfactors, details = extract_bfactors_directly_from_pdb(pdb_file_path)
    
    if direct_bfactors is None:
        print(f"❌ Failed to extract B-factors: {details}")
        continue
    
    print(f"Direct extraction B-factors shape: {direct_bfactors.shape}")
    
    # Compare lengths
    if len(direct_bfactors) != len(pickle_bfactors):
        print(f"⚠️  LENGTH MISMATCH:")
        print(f"   Direct extraction: {len(direct_bfactors)} residues")
        print(f"   Pickle file: {len(pickle_bfactors)} residues")
        print(f"   Sequence: {len(sequence)} amino acids")
        
        # Show first few residue details
        if isinstance(details, list) and len(details) > 0:
            print(f"   First 5 residues from PDB:")
            for j, detail in enumerate(details[:5]):
                print(f"     {detail['res_id']}: {detail['avg_bfactor']:.3f} (from {detail['num_atoms']} atoms)")
    else:
        print(f"✅ Lengths match: {len(direct_bfactors)} residues")
    
    # Compare values (only if lengths match or we can compare the overlapping part)
    compare_length = min(len(direct_bfactors), len(pickle_bfactors))
    direct_subset = direct_bfactors[:compare_length]
    pickle_subset = pickle_bfactors[:compare_length]
    
    # Calculate differences
    diff = np.abs(direct_subset - pickle_subset)
    max_diff = np.max(diff)
    mean_diff = np.mean(diff)
    
    print(f"\nB-factor comparison (first {compare_length} residues):")
    print(f"  Maximum absolute difference: {max_diff:.6f}")
    print(f"  Mean absolute difference: {mean_diff:.6f}")
    
    if max_diff < 1e-5:
        print(f"  ✅ PERFECT MATCH (within numerical precision)")
    elif max_diff < 0.001:
        print(f"  ✅ EXCELLENT MATCH (< 0.001 difference)")
    elif max_diff < 0.01:
        print(f"  ⚠️  GOOD MATCH (< 0.01 difference)")
    else:
        print(f"  ❌ SIGNIFICANT DIFFERENCES (> 0.01)")
        
        # Show details of differences
        large_diffs = np.where(diff > 0.01)[0]
        if len(large_diffs) > 0:
            print(f"  Residues with large differences:")
            for idx in large_diffs[:5]:  # Show first 5
                print(f"    Residue {idx}: Direct={direct_subset[idx]:.3f}, Pickle={pickle_subset[idx]:.3f}, Diff={diff[idx]:.3f}")
                if isinstance(details, list) and idx < len(details):
                    detail = details[idx]
                    print(f"      PDB details: {detail['res_id']}, atoms: {detail['atom_bfactors']}")
    
    # Show sample values for manual verification
    print(f"\nSample values (first 5 residues):")
    print(f"  Direct:  {direct_subset[:5]}")
    print(f"  Pickle:  {pickle_subset[:5]}")
    print(f"  Diff:    {diff[:5]}")
    
    # Show residue details for first few
    if isinstance(details, list):
        print(f"\nPer-atom breakdown (first 3 residues):")
        for j, detail in enumerate(details[:3]):
            print(f"  {detail['res_id']}: {detail['atom_bfactors']} → avg = {detail['avg_bfactor']:.3f}")

print(f"\n{'='*80}")
print("CORRECTNESS TEST SUMMARY")
print(f"{'='*80}")
print("If you see 'PERFECT MATCH' or 'EXCELLENT MATCH' for all test structures,")
print("then your pickle file B-factors are correctly extracted and averaged!")
print("Any significant differences would indicate a problem in the extraction process.")
```

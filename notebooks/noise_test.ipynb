```python
# Test the effect of dirichlet_multiplier_training on Dirichlet concentration parameters
import torch
import matplotlib.pyplot as plt
import numpy as np

def test_dirichlet_multiplier_effect():
    """Test how dirichlet_multiplier affects the Dirichlet distribution concentration."""
    
    K = 21  # Number of amino acids
    t_values = [1.0, 2.0, 4.0, 8.0]  # Different noise levels
    multipliers = [1.0, 2.0, 5.0, 10.0]  # Different multiplier values
    pos_true = 5  # Position of the true amino acid
    
    fig, axes = plt.subplots(len(t_values), len(multipliers), figsize=(20, 16))
    fig.suptitle('Effect of Dirichlet Multiplier on Concentration Parameters', fontsize=16)
    
    for i, t in enumerate(t_values):
        for j, multiplier in enumerate(multipliers):
            # Create one-hot encoding with true class at position pos_true
            y_onehot = torch.zeros(K)
            y_onehot[pos_true] = 1.0
            
            # Calculate alpha values as they would be in the updated sample_forward function
            alpha_true = (1.0 + t) * multiplier
            alpha_others = 1.0 * multiplier
            
            # Build full alpha vector
            alpha = torch.ones(K) * alpha_others
            alpha[pos_true] = alpha_true
            
            # Sample from Dirichlet
            dirichlet_dist = torch.distributions.Dirichlet(alpha)
            samples = dirichlet_dist.sample((1000,))  # Generate many samples
            mean_probs = samples.mean(dim=0)
            std_probs = samples.std(dim=0)
            
            # Plot results
            ax = axes[i, j]
            bars = ax.bar(range(K), mean_probs.numpy(), 
                         yerr=std_probs.numpy(), capsize=2, alpha=0.7)
            bars[pos_true].set_color('red')  # Highlight true position
            
            ax.set_title(f't={t}, mult={multiplier}\nTrue pos prob: {mean_probs[pos_true]:.3f}{std_probs[pos_true]:.3f}')
            ax.set_xlabel('Amino Acid Index')
            ax.set_ylabel('Mean Probability')
            ax.set_ylim(0, max(0.3, mean_probs.max().item() + 0.05))
            ax.axvline(x=pos_true, color='red', linestyle='--', alpha=0.5)
            
            # Add concentration values as text
            ax.text(0.02, 0.98, f'_true={alpha_true:.1f}\n_other={alpha_others:.1f}', 
                   transform=ax.transAxes, verticalalignment='top', fontsize=8,
                   bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
    
    plt.tight_layout()
    plt.show()
    
    # Summary statistics
    print("SUMMARY: Effect of Dirichlet Multiplier")
    print("="*50)
    for t in t_values:
        print(f"\nTime t = {t}:")
        for multiplier in multipliers:
            alpha_true = (1.0 + t) * multiplier
            alpha_others = 1.0 * multiplier
            concentration_ratio = alpha_true / alpha_others
            
            y_onehot = torch.zeros(K)
            y_onehot[pos_true] = 1.0
            alpha = torch.ones(K) * alpha_others
            alpha[pos_true] = alpha_true
            
            dirichlet_dist = torch.distributions.Dirichlet(alpha)
            samples = dirichlet_dist.sample((10000,))
            mean_true_prob = samples[:, pos_true].mean().item()
            
            print(f"  Multiplier {multiplier:4.1f}: _true={alpha_true:5.1f}, _other={alpha_others:4.1f}, "
                  f"ratio={concentration_ratio:4.1f}, P(true)={mean_true_prob:.3f}")

test_dirichlet_multiplier_effect()
```

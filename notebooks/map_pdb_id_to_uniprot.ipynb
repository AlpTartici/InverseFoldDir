```python
# load ../datasets/pdb_chain_uniprot.tsv.gz
import pandas as pd

# Read the gzipped TSV file
pdb_to_uniprot = pd.read_csv(
    "../datasets/pdb_chain_uniprot.tsv.gz",
    sep="\t",
    compression="gzip",
    header=1
)

# Quick check: show shape and first few rows
print(pdb_to_uniprot.shape)
print(pdb_to_uniprot.head())
```

```python
import json

file_path = '../datasets/cath-4.2/chain_set_splits.json'

with open(file_path, 'r') as file:
    data = json.load(file)

val_proteins = data['validation']
test_proteins = data['test']
```

```python
dict_pdb_to_protein = {}
dict_pdb_to_protein['validation'] = {}
dict_pdb_to_protein['test'] = {}
dict_missing = []
for protein in tqdm(val_proteins):
    [pdb, chain] = protein.split(".")
    if pdb in pdb_to_uniprot['PDB'].values:
        try:
            uniprot_id = pdb_to_uniprot[(pdb_to_uniprot['PDB'] == pdb) & (pdb_to_uniprot['CHAIN'] == chain)]['SP_PRIMARY'].values[0]
            dict_pdb_to_protein['validation'][protein] = uniprot_id
        except IndexError:
            dict_missing.append(protein)
    else:
        dict_missing.append(protein)

for protein in tqdm(test_proteins):
    [pdb, chain] = protein.split(".")
    if pdb in pdb_to_uniprot['PDB'].values:
        try:
            uniprot_id = pdb_to_uniprot[(pdb_to_uniprot['PDB'] == pdb) & (pdb_to_uniprot['CHAIN'] == chain)]['SP_PRIMARY'].values[0]
            dict_pdb_to_protein['test'][protein] = uniprot_id
        except IndexError:
            dict_missing.append(protein)
    else:
        dict_missing.append(protein)
```

```python
for pdb_id in data['validation']:
    coords, seq = data_pdb[pdb_id]['coords'], data_pdb[pdb_id]["seq"]
    len_seq = len(seq)
    if len_seq < 200 or len_seq > 300:
        continue
    # ensure there is no NaN in coords
    if any(np.isnan(coords[k]).any() for k in coords):
        continue
    # if any unknown amino acids in seq
    if any(aa not in 'ACDEFGHIKLMNPQRSTVWY' for aa in seq):
        continue
    print(f"{pdb_id}: {len_seq} residues")
```
